```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
print('libraries imported!')
```

```python
file = open('./Data/SMSSpamCollection.txt')
data = file.readlines()  # line by line 
file.close()
print("No. of email message : ",len(data) )# list of  lines
```

```python
print(data[2])
```

```python
x = []   # email message 
y = []   # labels 
for line in data:
    splitted_data = line.split('\t')
    if len(splitted_data) == 2: 
        y.append(splitted_data[0]) 
        x.append(splitted_data[1].replace('\n',''))
```

```python
len(x) ,len(y)
```

```python
df = pd.DataFrame({'email_messages':x,'labels':y})
df.head()
```

```python
# pd.read_table('./Data/SMSSpamCollection.txt')
```

```python
value_counts = df['labels'].value_counts()
```

```python
plt.bar(value_counts.index, value_counts.values, color='dodgerblue')
plt.title('Value Counts of Labels')
plt.show()
```

```python
plt.pie(value_counts.values,labels=value_counts.index,autopct='%2.3f%%')
plt.show()
```

```python
ham = df[df['labels']=='ham']
spam = df[df['labels']=='spam']
```

```python
spam.shape 
```

```python
ham.shape 
```

```python
HAM = ham.sample(1000)  # random sampling 
```

```python
HAM.shape 
```

```python
balance_df = pd.concat([spam,HAM])
```

```python
value_counts = balance_df['labels'].value_counts()
value_counts
```

```python
plt.pie(value_counts.values,labels=value_counts.index,autopct='%2.3f%%')
plt.show()
```

```python
balance_df.head()
```

```python
balance_df.reset_index(drop=True,inplace=True)  # new indexes 
```

```python
import re  # regular expression  
```

```python
balance_df['email_messages'][0]
```

```python
for i in range(len(balance_df)): 
    lower_message = balance_df['email_messages'][i].lower()
    clean_message = re.sub('[^a-zA-Z ]',"",lower_message)  # substitud  
    balance_df['email_messages'][i] = clean_message 
    
```

```python
balance_df['email_messages'][0]
```

```python
x  =  balance_df['email_messages'] 
y  =  balance_df['labels']
```

```python
X = x.values  # changed into numpy array 
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer  
## to convert text data to numerical representation 
```

```python
bow_obj =  CountVectorizer(max_features=1500,stop_words='english')
```

```python
# from nltk.corpus import stopwords 
```

```python
# len(stopwords.words('english')) 
```

```python
x_transformed = bow_obj.fit_transform(X)
x_transformed
```

```python
X.shape 
```

```python
X2 = x_transformed.toarray() 
```

```python
Y = y.map({'ham': 0, 'spam': 1})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X2,Y,test_size=0.10)
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB
```

```python
multNB = MultinomialNB()
berNB = BernoulliNB()
```

```python
multNB.fit(x_train,y_train)
berNB.fit(x_train,y_train)
print('You have succesfully trained algorithms')
```

```python
print('MultinomialNB')
print('Training Score : ',multNB.score(x_train,y_train)*100)
print('Test Score : ',multNB.score(x_test,y_test)*100)
print()
print('BernaulliNB')
print('Training Score : ',berNB.score(x_train,y_train)*100)
print('Test Score : ',berNB.score(x_test,y_test)*100)
```

```python
##Prediction
prediction = berNB.predict(x_test)
prediction
```

```python
np.array(y_test)
```

```python
evaluate_df = pd.DataFrame({'Actual': np.array(y_test),'prediction' : prediction})
evaluate_df.head(40)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
conf_matrix = confusion_matrix(evaluate_df['Actual'],evaluate_df['prediction'])  
conf_matrix
```

```python
print(classification_report(evaluate_df['Actual'],evaluate_df['prediction']))
```
